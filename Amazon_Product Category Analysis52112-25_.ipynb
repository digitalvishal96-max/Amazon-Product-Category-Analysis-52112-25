## Function to find all unique non-alphanumeric characters in a column

```python
import re

# Function to find all unique non-alphanumeric characters in a column
def find_special_chars(column):
    # Regex [^a-zA-Z0-9\s] means "everything except letters, numbers, and spaces"
    all_text = " ".join(column.astype(str))
    special_chars = set(re.findall(r'[^a-zA-Z0-9\s]', all_text))
    return special_chars

# Check the 'product_name' or 'about_product' columns
print("Special characters in Product Name:", find_special_chars(df['product_name']))
```

### That is quite a list! Seeing characters like \u200e (a hidden Left-to-Right mark), currency symbols like ₹, and various punctuation marks is very common in web-scraped Amazon data.

For a Recommendation System or Price Prediction model, you usually want to keep the text human-readable but remove "noise" that doesn't help the machine learn.

How to Clean These Characters
You should use a two-step approach: first, standardize common symbols, and second, strip out the remaining technical noise (like that \u200e code).

```python
import re
import html

def clean_product_text(text):
    if not isinstance(text, str):
        return ""
    
    # 1. Handle HTML entities (converts &amp; to &)
    text = html.unescape(text)
    
    # 2. Standardize some symbols before stripping others
    text = text.replace('₹', 'Rupees ').replace('&', ' and ')
    
    # 3. Remove technical noise like \u200e and other non-printable characters
    # This regex keeps letters, numbers, and standard punctuation
    text = re.sub(r'[^\x00-\x7F]+', ' ', text) 
    
    # 4. Remove the specific characters you found that are just "noise"
    # We'll keep spaces and basic alphanumeric characters
    text = re.sub(r'[*°\]@()~™、:”’–;\[…\'!,，|_+="<%>®]', ' ', text)
    
    # 5. Clean up extra whitespace created by the removals
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Apply to your columns
df['product_name_clean'] = df['product_name'].apply(clean_product_text)
df['about_product_clean'] = df['about_product'].apply(clean_product_text)
```

However, for a Machine Learning model, too many special characters can create "noise" that makes it harder for the algorithm to find actual patterns

```python
import re

def final_clean(text):
    # 1. Standardize technical abbreviations
    text = text.replace('"', ' inch').replace("'", ' feet')
    
    # 2. Remove purely decorative symbols like 【 】, ™, ®, etc.
    text = re.sub(r'[【】™®©]', ' ', text)
    
    # 3. Strip hidden Unicode characters (like \u200e)
    text = text.encode("ascii", "ignore").decode()
    
    # 4. Remove excessive punctuation but keep numbers and letters
    text = re.sub(r'[^a-zA-Z0-9\s.-]', ' ', text)
    
    # 5. Collapse multiple spaces into one
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['about_product_final'] = df['about_product'].apply(final_clean)
```

In data science, creating these "extra" columns is a process called **Feature Engineering**. Since your original data consists of raw, unformatted text, these columns act as refined versions of that data, making it readable for machine learning algorithms.

Here is why each specific column was created and how they help your analysis:

### **1. special_char_count**

This column serves as a **Diagnostic Metric**.

* **Why it’s created:** To quantify the "noise" or complexity of the text in each row.
* **How it’s useful:** It helps you identify rows that might be "dirty" or contain non-standard formatting (like hidden Unicode or excessive symbols). High counts often flag technical products (like printers) where measurements and symbols are dense, allowing you to tailor your cleaning strategy for different product types.

### **2. product_name_clean & about_product_clean**

These are **Intermediate Standardized** columns.

* **Why they’re created:** To strip away common formatting artifacts like HTML entities (`&amp;`) and currency symbols (`₹`) while keeping the core text intact.
* **How it’s useful:** They provide a "safe" version of your data for basic exploration. By keeping these separate from the raw columns, you can always go back and check if your cleaning process accidentally deleted important words.

### **3. about_product_final**

This is your **Model-Ready Feature**.

* **Why it’s created:** This is the ultra-clean version of your text where all non-essential "trash" (like `\u200e` or decorative icons) has been removed, and technical symbols have been converted to words (e.g., `"` to `inch`).
* **How it’s useful:** This is the column you will actually feed into your **Vectorization** process (like TF-IDF). Because it is highly consistent, the machine learning model won't get "distracted" by punctuation or hidden symbols, leading to much higher accuracy in your **Recommendation Engine** or **Price Prediction** model.

---

### **The Overall Benefit to Your Project**

By maintaining these separate columns instead of overwriting your original data, you create a **transparent workflow**. This allows you to:

1. **Debug Faster:** If the model makes a strange recommendation, you can trace the text through its different stages of cleaning to find the error.
2. **Improve Accuracy:** Models perform better on "normalized" text. For example, a model might see `12"` and `12 inch` as different things, but after your cleaning, it treats them as the exact same feature.
3. **Save Time:** Once you've confirmed `about_product_final` is perfect, you can drop the other helper columns to save memory during the heavy training phase.

Since machines cannot understand words, Vectorization is the process of converting your cleaned text into a numerical matrix. For a beginner, the most effective method is TF-IDF (Term Frequency-Inverse Document Frequency).

Why use TF-IDF?
Unlike a simple word count, TF-IDF penalizes words that appear everywhere (like "the" or "product") and rewards unique, descriptive words (like "Bluetooth," "Waterproof," or "Ergonomic"). This ensures your model focuses on the important features of the Amazon products.

The Vectorization Step
Here is how you turn your about_product_final column into a mathematical matrix:

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Initialize the Vectorizer
# stop_words='english' removes common words like 'and', 'the', etc.
tfidf = TfidfVectorizer(stop_words='english', max_features=7000)

# 2. Fit and Transform the cleaned text
# This creates a sparse matrix of numbers
tfidf_matrix = tfidf.fit_transform(df['about_product_final'])

# 3. Check the shape
print(f"Matrix Shape: {tfidf_matrix.shape}")
# Example Output: (1465, 5000) -> 1,465 products described by 5,000 unique words
```

That matrix shape (1465, 7000) is exactly what you want to see! It tells a very specific story about your data:

1465: This matches your total number of products. Each row represents one unique item.

5000: This is the "vocabulary" your model created. It means that across all 1,465 products, the vectorizer identified the 5,000 most important words (features) to describe them.

The Final Step: Cosine Similarity
Now that your text is numerical, we use Cosine Similarity to calculate the mathematical "closeness" between products. This is what powers the "People also viewed" or "Similar items" section on Amazon.

Cosine Similarity looks at the angle between two vectors. If two products both have high values for "Wireless," "Bluetooth," and "Headphones," the angle between them will be very small, resulting in a similarity score close to 1.0.

The Python Implementation
Here is the code to find the most similar products to any item in your dataset:

```python
from sklearn.metrics.pairwise import cosine_similarity

# 1. Calculate the similarity between all products
# This creates a 1465 x 1465 matrix
similarity_matrix = cosine_similarity(tfidf_matrix)

# 2. Create a function to get recommendations
def get_recommendations(product_name, df, similarity_matrix):
    # Find the index of the product that matches the name
    idx = df[df['product_name'] == product_name].index[0]
    
    # Get the pairwise similarity scores for all products with that product
    sim_scores = list(enumerate(similarity_matrix[idx]))
    
    # Sort products based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the scores of the 5 most similar products (excluding itself)
    sim_scores = sim_scores[1:6]
    
    # Get the product indices
    product_indices = [i[0] for i in sim_scores]
    
    # Return the top 5 most similar products
    return df['product_name'].iloc[product_indices]

# Try it out!
print(get_recommendations(df['product_name'].iloc[0], df, similarity_matrix))
```

### **Analysis of the "Confusion" Categories**

Based on your results, the boundary blur is caused by four distinct types of linguistic noise:

* **Generic Technical Attributes:** Words like **usb**, **power**, **battery**, **charge**, and **type** are extremely common across almost all modern electronics. Because these appear in everything from an iPhone to a pedestal fan, the model thinks those items are related.
* **Retail & Marketing "Fluff":** Terms like **premium**, **design**, **free**, **colour**, and **purpose** are "marketing speak" that add zero value to the actual product category but carry high mathematical weight if they appear frequently in descriptions.
* **Shared Physical Units:** The numbers **10** and **14** (referring to inches, watts, or volts) are causing products of the same *size* or *power* to be grouped together even if they do different things (e.g., a 10-inch filter vs. a 10-inch screen).

### The "Elbow Method"
On its own, a single inertia score doesn't tell you if 10 was the "perfect" number of clusters. To find the optimal number, we calculate inertia for a range of clusters and look for the "Elbow" in the graph—the point where adding more clusters no longer significantly reduces inertia.

```python
# Calculate inertia for different values of K
inertias = []
K_range = range(1, 21)

for k in K_range:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(tfidf_matrix)
    inertias.append(model.inertia_)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertias, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

To definitively pick the best version of your model, we use the Silhouette Score to measure how well-separated your clusters are. While the Inertia score you calculated (1295.34) tells you how "tight" the clusters are, the Silhouette Score tells you if the boundaries between categories like "Charging Cables" and "Smartwatches" are distinct or blurry.

1. How the Silhouette Score Works
The score is calculated for each product by comparing:

Cohesion: How close the product is to others in its own cluster.

Separation: How far the product is from the nearest neighboring cluster.

A higher score (closer to 1) means your products are perfectly matched to their categories, while a score near 0 means your categories are overlapping.

2. Code to Compare k=8, 10, and 12
This script iterates through your three target values, fits the model, and prints the score for each so you can see which one "peaks."

# K-Means Algorithm

This code is a classic example of Hyperparameter Tuning. Specifically, it is searching for the optimal number of clusters ($k$) for a K-Means model by using the Silhouette Score as the evaluation metric.Here is the breakdown of what each part is doing:

1. The Core LibrariesKMeans: The algorithm that groups your data (the tfidf_matrix) into $k$ different groups based on distance.silhouette_score:

This is the "judge." It measures how similar an object is to its own cluster compared to other clusters.Score range: -1 to +1.Interpretation: A high score (closer to 1) means the clusters are well-separated and dense. A score near 0 indicates overlapping clusters.

2. The Logic LoopThe code iterates through your list of k_values (8, 10, and 12):fit_predict: It trains the model on your data and immediately assigns a cluster label (e.g., "Cluster 0", "Cluster 1") to every row in your matrix.Scoring: It feeds the raw data (tfidf_matrix) and the resulting labels into the silhouette function to see how "clean" the grouping was.Storage: It saves the score in a dictionary (results) where the key is $k$ and the value is the score.


3. Identifying the "Winner"This line looks through your results dictionary and finds the key (the number of clusters) that has the highest silhouette score. In clustering, the highest score usually represents the most distinct and logical grouping of your data.

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

# Define the k-values we want to test
k_values = [8, 10, 12]
results = {}

for k in k_values:
    # Initialize and fit the model
    temp_km = KMeans(n_clusters=k, random_state=42)
    labels = temp_km.fit_predict(tfidf_matrix)
    
    # Calculate the Silhouette Score
    score = silhouette_score(tfidf_matrix, labels)
    results[k] = score
    print(f"k={k} | Silhouette Score: {score:.4f}")

# Find the winner
best_k = max(results, key=results.get)
print(f"\nDefinitive Best Version: k={best_k}")
```

Here is the line-by-line breakdown of what is happening mathematically:

1. The Input: description
The function takes a raw string (like your "Fast charging..." example). At this stage, the computer just sees a sequence of characters; it has no numerical meaning yet.

2. tfidf.transform([description])
This is the most critical step.

The "Dictionary" Rule: You use the tfidf vectorizer that was already "fitted" on your 1,465 products. It uses the exact same 5,000-word vocabulary it built earlier.

The Conversion: It looks for words like "charging," "braided," and "nylon" in its dictionary. It then creates a sparse vector (a row of 5,000 numbers) where most values are 0, but the positions corresponding to those keywords get a specific mathematical weight.

Why transform and not fit_transform? You only want to transform the new text based on the old rules. If you "fit" again, the model would change its vocabulary, and the numbers wouldn't match your clusters anymore.

3. km.predict(new_vec)[0]
Now that the text is a vector of numbers, it can be plotted in your 7,000-dimensional space.

Finding the Center: The K-Means model (km) looks at where this new point lands and calculates which of the 12 cluster centers (centroids) is closest.

The Result: km.predict returns an array (e.g., [5]). We add [0] at the end to just grab the integer 5 itself.

4. The Output: return cluster
The function hands back the cluster ID. If Cluster 6 was your "Charging Cables" group, you now know exactly how to categorize this new item.

### To find the top 5 most similar items within a cluster, we use Cosine Similarity. Since K-Means already grouped similar items together, we now calculate the exact mathematical "closeness" between your new product and every other product inside that specific group.

The Logic of Recommendation
Filter: We only look at products that share the same cluster_label.

Measure: We calculate the distance between the TF-IDF vector of your input and the vectors of the items in that cluster.

Sort: We rank them from highest similarity (1.0) to lowest.

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_top_5(description, df, tfidf, tfidf_matrix):
    # 1. Vectorize the input description
    query_vec = tfidf.transform([description])
    
    # 2. Find which cluster this description belongs to
    predicted_cluster = km.predict(query_vec)[0]
    
    # 3. Filter the main DataFrame to only include items from that cluster
    cluster_indices = df[df['cluster_label'] == predicted_cluster].index
    cluster_tfidf_matrix = tfidf_matrix[cluster_indices]
    
    # 4. Calculate similarity between the query and everything in the cluster
    similarities = cosine_similarity(query_vec, cluster_tfidf_matrix).flatten()
    
    # 5. Get the indices of the top 5 most similar items
    # argsort gives indices of sorted values; [-5:] gets the last 5 (highest)
    top_5_indices = similarities.argsort()[-5:][::-1]
    
    # 6. Map back to the original DataFrame to get product names/details
    recommended_items = df.iloc[cluster_indices[top_5_indices]]
    
    return recommended_items[['product_name', 'category']]

# Example usage
results = recommend_top_5("Fast charging braided cable", df, tfidf, tfidf_matrix)
print(results)
```
